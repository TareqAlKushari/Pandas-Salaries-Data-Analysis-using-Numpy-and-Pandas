```python
mean_basePay=data['BasePay'].mean()
mean_basePay
```

```python
data["BasePay"].fillna(mean_basePay,inplace=True)
```

```python
data.isnull().sum()
```

```python
data["OvertimePay"].fillna(0 ,inplace=True)
data["OtherPay"].fillna(0 ,inplace=True)
data["Benefits"].fillna(0 ,inplace=True)
```

```python
data.isnull().sum()
```

```python
data["Notes"].fillna("Not have notes",inplace=True)
data["Status"].fillna("Empty",inplace=True)
```

```python
data.isnull().sum()
```

```python
data["TotalPay"]=data["BasePay"]+data["OvertimePay"]+data["OtherPay"]
```

```python
data["TotalPayBenefits"]=data["TotalPay"]+data["Benefits"]
```
